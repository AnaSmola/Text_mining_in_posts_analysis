```python
#import libraries

import pandas as pd
import string
import re
import numpy as np

#upload the file

data = pd.read_excel(r'C:\Users\dell\Desktop\pliki\testfresh.xlsx')
data.head(20)
```

```python
#check the type of data
data.info()
```

```python
# change empty values to "NAN"
data[['post','dt 2']] = data[['post','dt 2']].replace(np.nan, '', regex=True)
```

```python
data
```

```python
#change the data types to string
data["postfs"] = data["post"].astype(str) +""+ data["dt 2"].astype(str)
```

```python
#drop unnecessary columns 
data.drop(['post', 'dt 2'], axis=1, inplace=True)
```

```python
data
```

```python
data['comments'] = data['comments'].astype(str)
```

```python
#delete text from column 'comments'
p = re.compile(r'\D')
data['comments'] = [p.sub('', x) for x in data['comments']]
```

```python
data
```

```python
data.info()
```

```python
#change empty values "NAN" to zero in column 'likes'

data['likes']  = data['likes'].replace(np.nan, 0, regex=True)
```

```python
#change floats to int in column 'likes'
data['likes'] = data['likes'].astype(int)
```

```python
#change NAN values in column 'comments' to 0
data['comments']  = pd.to_numeric(data['comments'] )
```
